## Model component behavior

### Subtask:
Develop mathematical or rule-based models for each component to simulate their energy consumption, generation, and storage over time.


**Reasoning**:
Define classes for each grid component with parameters and methods to simulate their behavior over time.



```python
class Component:
    def __init__(self, name):
        self.name = name

    def get_energy_flow(self, time_step):
        """Abstract method to get energy flow (consumption or generation) at a given time step."""
        raise NotImplementedError

class ConsumptionComponent(Component):
    def __init__(self, name, base_consumption, daily_profile=None, variability=0.1):
        super().__init__(name)
        self.base_consumption = base_consumption
        self.daily_profile = daily_profile if daily_profile is not None else np.ones(24) # Default to constant consumption
        self.variability = variability

    def get_energy_flow(self, time_step):
        hour_of_day = time_step % 24
        profile_factor = self.daily_profile[hour_of_day]
        random_variation = np.random.normal(1, self.variability)
        consumption = self.base_consumption * profile_factor * random_variation
        return -max(0, consumption) # Consumption is negative energy flow

class GenerationComponent(Component):
    def __init__(self, name, capacity, availability=1.0, variability=0.05): # Try and play with the availability
        super().__init__(name)
        self.capacity = capacity
        self.availability = availability
        self.variability = variability

    def get_energy_flow(self, time_step):
        random_variation = np.random.normal(1, self.variability)
        generation = self.capacity * self.availability * random_variation
        return max(0, generation) # Generation is positive energy flow

class StorageComponent(Component):
    def __init__(self, name, capacity, charge_efficiency=0.9, discharge_efficiency=0.9, max_charge_rate=None, max_discharge_rate=None, initial_charge=0):
        super().__init__(name)
        self.capacity = capacity
        self.charge_efficiency = charge_efficiency
        self.discharge_efficiency = discharge_efficiency
        self.max_charge_rate = max_charge_rate if max_charge_rate is not None else capacity
        self.max_discharge_rate = max_discharge_rate if max_discharge_rate is not None else capacity
        self.current_charge = min(capacity, max(0, initial_charge))

    def get_energy_flow(self, time_step, desired_flow):
        """Calculates actual energy flow based on desired flow and battery constraints."""
        if desired_flow > 0: # Try to discharge
            discharge_amount = min(desired_flow / self.discharge_efficiency, self.current_charge, self.max_discharge_rate)
            return discharge_amount * self.discharge_efficiency # Actual energy delivered
        elif desired_flow < 0: # Try to charge, Decreasing the desired flow is the optimization
            charge_amount = min(-desired_flow * self.charge_efficiency, self.capacity - self.current_charge, self.max_charge_rate)
            self.current_charge += charge_amount
            return -charge_amount / self.charge_efficiency # Actual energy drawn
        else:
            return 0

    def get_state_of_charge(self):
        return self.current_charge / self.capacity * 100

# Example daily profile for houses (higher in morning and evening)
house_daily_profile = np.array([0.8, 0.7, 0.6, 0.5, 0.5, 0.6, 0.8, 1.2, 1.5, 1.3, 1.0, 0.9, 1.0, 1.1, 1.2, 1.3, 1.5, 1.8, 2.0, 1.8, 1.5, 1.2, 1.0, 0.9])


houses = ConsumptionComponent("Houses", base_consumption=50, daily_profile=house_daily_profile, variability=0.15)
hospital = ConsumptionComponent("Hospital", base_consumption=200, variability=0.05) # More stable consumption
water_treatment = ConsumptionComponent("Water Treatment", base_consumption=150, variability=0.03) # Even more stable
generator = GenerationComponent("Generator", capacity=500, availability=0.95, variability=0.02)
battery = StorageComponent("Battery", capacity=1000, max_charge_rate=200, max_discharge_rate=200, initial_charge=500)

# Demonstrate getting energy flow for a few time steps
time_steps = 5
for t in range(time_steps):
    house_demand = houses.get_energy_flow(t)
    hospital_demand = hospital.get_energy_flow(t)
    water_treatment_demand = water_treatment.get_energy_flow(t)
    generator_output = generator.get_energy_flow(t)

    print(f"Time Step {t}:")
    print(f"  Houses Demand: {house_demand:.2f} kWh")
    print(f"  Hospital Demand: {hospital_demand:.2f} kWh")
    print(f"  Water Treatment Demand: {water_treatment_demand:.2f} kWh")
    print(f"  Generator Output: {generator_output:.2f} kWh")

    # Example of battery interaction (simplified for demonstration)
    net_energy = generator_output + house_demand + hospital_demand + water_treatment_demand
    battery_flow = battery.get_energy_flow(t, -net_energy) # Try to balance the grid
    print(f"  Battery Flow: {battery_flow:.2f} kWh")
    print(f"  Battery State of Charge: {battery.get_state_of_charge():.2f}%")
```

## Integrate components

### Subtask:
Combine the models of individual components to create a holistic simulation of the electric grid's behavior.


**Reasoning**:
Combine the component models into a holistic simulation by iterating through time steps, calculating total demand and generation, determining net energy, and updating the battery state. Store the results in a DataFrame.



```python
components = {
    "Houses": houses,
    "Hospital": hospital,
    "Water Treatment": water_treatment,
    "Generator": generator,
    "Battery": battery
}

simulation_results = []
time_steps = 100

for t in range(time_steps):
    # 3. Calculate the total demand from all consumption components.
    total_demand = sum(comp.get_energy_flow(t) for name, comp in components.items() if isinstance(comp, ConsumptionComponent))

    # 4. Calculate the total generation from all generation components.
    total_generation = sum(comp.get_energy_flow(t) for name, comp in components.items() if isinstance(comp, GenerationComponent))

    # 5. Calculate the net energy flow (generation - demand).
    net_energy = total_generation + total_demand # demand is negative

    # 6. Determine how the battery should respond to the net energy flow
    # The battery tries to balance the net energy flow.
    desired_battery_flow = -net_energy

    # 7. Update the battery's state of charge
    battery_flow = battery.get_energy_flow(t, desired_battery_flow)

    # Record the energy flow of each component and the battery's state of charge
    time_step_results = {
        "Time Step": t,
        "Houses Energy Flow": components["Houses"].get_energy_flow(t),
        "Hospital Energy Flow": components["Hospital"].get_energy_flow(t),
        "Water Treatment Energy Flow": components["Water Treatment"].get_energy_flow(t),
        "Generator Energy Flow": components["Generator"].get_energy_flow(t),
        "Battery Energy Flow": battery_flow,
        "Net Energy Before Battery": net_energy,
        "Net Energy After Battery": net_energy + battery_flow, # Should be close to zero if battery can balance
        "Battery State of Charge (%)": battery.get_state_of_charge()
    }
    simulation_results.append(time_step_results)

# Convert results to a pandas DataFrame
simulation_df = pd.DataFrame(simulation_results)

# Display the first few rows of the results DataFrame
display(simulation_df.head())
```

## Modify the simulation environment

### Subtask:
Adapt your existing simulation code to function as an OpenAI Gym-like environment. This involves creating a class with `reset()` and `step()` methods.


**Reasoning**:
Adapt the existing simulation code into an OpenAI Gym-like environment by creating a class with `reset()` and `step()` methods, defining the observation and action spaces, initializing components, implementing the reset logic, and implementing the step logic to process actions, advance the simulation, calculate rewards, and return the next state, reward, done flag, and info.



```python
import gymnasium as gym
from gymnasium import spaces

class GridEnv(gym.Env):
    def __init__(self, num_time_steps=100, initial_battery_charge=500):
        super(GridEnv, self).__init__()

        self.num_time_steps = num_time_steps
        self.initial_battery_charge = initial_battery_charge
        self.current_time_step = 0

        # Define the components.
        self.houses = ConsumptionComponent("Houses", base_consumption=50, daily_profile=house_daily_profile, variability=0.15)
        self.hospital = ConsumptionComponent("Hospital", base_consumption=200, variability=0.05)
        self.water_treatment = ConsumptionComponent("Water Treatment", base_consumption=150, variability=0.03)
        self.generator = GenerationComponent("Generator", capacity=500, availability=0.90, variability=0.03)
        self.battery = StorageComponent("Battery", capacity=2000, max_charge_rate=500, max_discharge_rate=500, initial_charge=self.initial_battery_charge)


        self.components = {
            "Houses": self.houses,
            "Hospital": self.hospital,
            "Water Treatment": self.water_treatment,
            "Generator": self.generator,
            "Battery": self.battery
        }


        # Define observation space
        # Houses, Hospital, Water Treatment, Generator, Battery SOC, Time of Day
        # Scale observation space to improve training stability
        low_obs = np.array([
            -500., # Houses Energy Flow (scaled based on expected range)
            -500., # Hospital Energy Flow (scaled based on expected range)
            -500., # Water Treatment Energy Flow (scaled based on expected range)
            0.,       # Generator Energy Flow (scaled based on expected range)
            0.,       # Battery State of Charge (%) (0-100)
            0.        # Time of Day (0-23)
        ], dtype=np.float32)

        high_obs = np.array([
            500., # Houses Energy Flow (scaled based on expected range)
            500., # Hospital Energy Flow (scaled based on expected range)
            500., # Water Treatment Energy Flow (scaled based on expected range)
            1000., # Generator Energy Flow (Increased upper bound)
            100.,    # Battery State of Charge (%)
            23.      # Time of Day
        ], dtype=np.float32)

        self.observation_space = spaces.Box(low=low_obs, high=high_obs, dtype=np.float32)

        # Define action space
        # Battery Desired Energy Flow (continuous value within battery limits)
        low_action = np.array([-self.battery.max_charge_rate], dtype=np.float32)
        high_action = np.array([self.battery.max_discharge_rate], dtype=np.float32)
        self.action_space = spaces.Box(low=low_action, high=high_action, dtype=np.float32)

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.current_time_step = 0
        self.battery.current_charge = self.initial_battery_charge # Reset battery charge
        self.last_generator_flow = self.generator.get_energy_flow(self.current_time_step) # Initialize last generator flow


        # Get the initial observation
        obs = self._get_observation()
        info = {} # Add info dictionary for gymnasium compatibility
        return obs, info

    def step(self, action):
        # Apply the action to the battery
        desired_battery_flow = action[0] # Action is a single value array

        # Calculate energy flows for the current time step before battery interaction
        house_flow = self.houses.get_energy_flow(self.current_time_step)
        hospital_flow = self.hospital.get_energy_flow(self.current_time_step)
        water_treatment_flow = self.water_treatment.get_energy_flow(self.current_time_step)
        generator_flow = self.generator.get_energy_flow(self.current_time_step)

        # Calculate net energy before battery
        net_energy_before_battery = generator_flow + house_flow + hospital_flow + water_treatment_flow

        # Get the actual battery energy flow based on the desired flow and battery constraints
        # The battery's internal state (current_charge) is updated within get_energy_flow
        battery_flow = self.battery.get_energy_flow(self.current_time_step, desired_battery_flow)

        # Calculate net energy after battery
        net_energy_after_battery = net_energy_before_battery + battery_flow

        # Calculate the reward
        reward = self._calculate_reward(generator_flow, self.battery.get_state_of_charge(), net_energy_after_battery, self.current_time_step, self.last_generator_flow, generator_flow, battery_flow, net_energy_before_battery)

        # Update last generator flow for the next step's reward calculation
        self.last_generator_flow = generator_flow

        # Advance the time step
        self.current_time_step += 1

        # Determine if the episode is done
        terminated = self.current_time_step >= self.num_time_steps
        truncated = False # Assuming no truncation in this environment for now

        # Get the next observation
        obs = self._get_observation()

        # Optional info dictionary
        info = {
            'houses_flow': house_flow,
            'hospital_flow': hospital_flow,
            'water_treatment_flow': water_treatment_flow,
            'generator_flow': generator_flow,
            'battery_flow': battery_flow,
            'net_energy_before_battery': net_energy_before_battery,
            'net_energy_after_battery': net_energy_after_battery,
            'battery_soc': self.battery.get_state_of_charge() # SOC *after* the step
        }


        return obs, reward, terminated, truncated, info # Return terminated and truncated for gymnasium

    def _get_observation(self):
        # Construct the observation vector
        # Scale energy flows to keep values within a reasonable range
        house_flow = self.houses.get_energy_flow(self.current_time_step)
        hospital_flow = self.hospital.get_energy_flow(self.current_time_step)
        water_treatment_flow = self.water_treatment.get_energy_flow(self.current_time_step)
        generator_flow = self.generator.get_energy_flow(self.current_time_step)
        battery_soc = self.battery.get_state_of_charge()
        time_of_day = self.current_time_step % 24

        obs = np.array([
            house_flow,
            hospital_flow,
            water_treatment_flow,
            generator_flow,
            battery_soc,
            time_of_day
        ], dtype=np.float32)

        return obs

    def _calculate_reward(self, generator_flow, battery_soc, net_energy_after_battery, time_step, last_generator_flow, current_generator_flow, battery_flow, net_energy_before_battery):
        reward = 0

        # --- Reward Weights (Adjust these based on desired priorities) ---
        W_IMBALANCE = 5.0  # Increased weight: strongly prioritize meeting demand
        W_LOW_SOC = 4.0   # Increased weight: ensure resilience
        W_SMOOTH = 1.0    # Slightly increased weight for smoothing
        W_CHARGE = 0.5    # Reward for charging
        W_SOC_TARGET = 0.2 # Reward for being in target SOC range

        # --- Penalties ---
        # Penalty for significant net energy imbalance after battery intervention (power quality/grid stability)
        reward -= abs(net_energy_after_battery) * W_IMBALANCE  # Linear penalty

        # Penalty for high generator usage (discourage relying solely on the generator)
        # Penalize generator output relative to its capacity
        reward -= (abs(generator_flow) / self.generator.capacity) * W_IMBALANCE

        # Penalty for large fluctuations in generator output (encourage smoothing)
        generator_change = abs(current_generator_flow - last_generator_flow)
        reward -= (generator_change / self.generator.capacity) * 7.0 * W_SMOOTH

        # Penalty for low battery state of charge (risk of not being able to meet critical demand)
        if battery_soc < 20:
            low_soc_penalty = -((20 - battery_soc) / 20.0)**2 # Quadratic penalty for being below 20%
            reward += low_soc_penalty * W_LOW_SOC
        elif battery_soc > 80:
             high_soc_penalty = -((battery_soc - 80) / 20.0)**2 # Quadratic penalty for being above 80%
             reward += high_soc_penalty * W_LOW_SOC

        # --- Rewards ---
        # Reward for charging the battery (store excess energy) - only reward if there is excess generation or need
        # Only reward charging if net_energy_before_battery is positive (excess generation) or battery_soc is low
        if battery_flow < 0 and (net_energy_before_battery > 0 or battery_soc < 50):
             reward += abs(battery_flow) * W_CHARGE # Reward based on amount charged

        # Reward for keeping battery SOC within a target range (e.g., 30% to 70%) - encourages usability
        if 30 <= battery_soc <= 70:
            reward += W_SOC_TARGET # Small positive reward for being in the preferred range


        return reward / 100.0 # Adjust overall reward scale



# Example usage (optional, for testing the environment)
env = GridEnv(num_time_steps=24)
obs, info = env.reset()
print("Initial Observation:", obs)

# # Example of taking a random action
action = env.action_space.sample()
print("Sample Action:", action)

obs, reward, terminated, truncated, info = env.step(action)
print("Step 1 Results:")
print("Observation:", obs)
print("Reward:", reward)
print("Terminated:", terminated)
print("Truncated:", truncated)
print("Info:", info)
```
